# Food Delivery Hackathon Analysis
This notebook contains my submission for the Food Delivery Hackathon.  
I merged three datasets (`orders.csv`, `users.json`, `restaurants.sql`) and performed analysis to answer the hackathon questions.


## Step 1: Load Datasets
We load the three datasets into Pandas:
- Orders (CSV)
- Users (JSON)
- Restaurants (SQL)


```python
import pandas as pd
orders = pd.read_csv("orders.csv")
print(orders.head())
```

```python
users = pd.read_json("users.json")
print(users.head())
```

```python
import sqlite3

conn = sqlite3.connect(":memory:")
with open("restaurants.sql", "r") as f:
    sql_script = f.read()
conn.executescript(sql_script)

restaurants = pd.read_sql_query("SELECT * FROM restaurants;", conn)
print(restaurants.head())
```

## Step 2: Merge Datasets
We merge the datasets on `user_id` and `restaurant_id` to create a unified dataset.


```python
merged = orders.merge(users, on="user_id", how="left")
merged = merged.merge(restaurants, on="restaurant_id", how="left")
print(merged.head())
```
